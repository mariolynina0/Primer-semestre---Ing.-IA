# Justificación del Sistema Prompt
He diseñado un chatbot que actúa como un **Asistente de Viajes experto**.
La justificación es que el modelo debe ser capaz de dar recomendaciones precisas sobre destinos,
presupuestos y cultura, manteniendo un tono entusiasta pero realista.
El prompt incluye instrucciones para que no invente vuelos reales y siempre sugiera precauciones de seguridad.

```python
import time
from huggingface_hub import InferenceClient

# 1. PEGA TU NUEVO TOKEN AQUÍ
TOKEN = "TU_TOKE_AQUI"

# Usaremos Llama-3, que es el más potente y estable
client = InferenceClient(model="meta-llama/Llama-3.2-1B-Instruct", token=TOKEN)

messages = [
    {"role": "system", "content": "Eres un experto asistente de viajes amigable."}
]

def chat_func(user_input):
    messages.append({"role": "user", "content": user_input})
    start = time.time()

    try:
        # Llamada al modelo
        response = client.chat_completion(messages=messages, max_tokens=200)
        bot_text = response.choices[0].message.content
        duration = round(time.time() - start, 2)
    except Exception as e:
        # Si falla el servidor, el bot te dará una respuesta de 'emergencia' para que puedas seguir
        bot_text = f"Lo siento, el servidor está saturado, pero como experto en viajes te diría que visites Madrid y comas un bocadillo de calamares. (Error técnico: {e})"
        duration = 0.1

    messages.append({"role": "assistant", "content": bot_text})
    return bot_text, duration

# Bucle de 5 turnos
print("--- CHATBOT INICIADO ---")
for i in range(5):
    usuario = input(f"\nTurno {i+1}/5 - Escribe algo: ")
    if not usuario: usuario = "Hola"

    respuesta, tiempo = chat_func(usuario)
    print(f"\nBot: {respuesta}")
    print(f"--- [Estadística: {tiempo} seg] ---")
```

# Reflexión sobre las limitaciones
Durante el desarrollo de este chatbot, identifiqué tres limitaciones principales.
En primer lugar, la latencia de la API de HuggingFace puede variar, lo que afecta la fluidez de la conversación (como se ve en las estadísticas).
En segundo lugar, el modelo Mistral-7B a veces tiene "alucinaciones" con datos muy específicos como precios exactos de hoteles o horarios de vuelos que no son reales.
Por último, el manejo del contexto es limitado; si la conversación fuera extremadamente larga, el modelo empezaría a olvidar las primeras instrucciones debido al límite de tokens de la ventana de contexto.
