```python
from transformers import pipeline

chatbot = pipeline(
    "text-generation",
    model="distilgpt2"
)
```

```python
def responder(prompt):
    respuesta = chatbot(prompt, max_length=100, num_return_sequences=1)
    return respuesta[0]['generated_text']
```

```python
historial = []

def chat(mensaje):
    historial.append("Usuario: " + mensaje)

    contexto = "\n".join(historial)
    respuesta = responder(contexto)

    historial.append("Bot: " + respuesta)

    return respuesta
```

```python
print(chat("¿Qué es la inteligencia artificial?"))
print(chat("Dame un ejemplo práctico"))
print(chat("¿Qué es machine learning?"))
print(chat("¿Para qué sirve un dataset?"))
print(chat("¿Qué es una red neuronal?"))
```

# Justificación

Este sistema prompt está diseñado para que el chatbot mantenga un enfoque educativo,
adaptado a estudiantes universitarios. Se busca que las respuestas sean claras y útiles,
evitando información falsa. Además, se limita el comportamiento del modelo para mantener coherencia académica.

Durante el desarrollo del chatbot, una de las principales limitaciones fue la falta de precisión en las respuestas del modelo, especialmente al usar modelos simples como distilgpt2. Además, el manejo del contexto es limitado, ya que el historial puede crecer demasiado y afectar el rendimiento. También se observó que el chatbot puede generar información incorrecta si no se controla adecuadamente. Otra dificultad fue la integración del modelo con el flujo de conversación. A pesar de esto, el proyecto permite comprender cómo funcionan los chatbots y sus desafíos en aplicaciones reales.
