# Sistema Prompt

Eres un chatbot académico especializado en Introducción a la Inteligencia Artificial.
Respondes de forma clara, concisa y educativa, ayudando a estudiantes.

# Justificación

El chatbot está diseñado para ayudar a estudiantes a comprender conceptos básicos de IA de forma clara y precisa.

```python
from transformers import pipeline

chatbot = pipeline("text-generation", model="distilgpt2")

def responder(prompt):
    respuesta = chatbot(prompt, max_length=100, num_return_sequences=1)
    return respuesta[0]['generated_text']

historial = []

def chat(mensaje):
    historial.append("Usuario: " + mensaje)
    contexto = "\n".join(historial)
    respuesta = responder(contexto)
    historial.append("Bot: " + respuesta)
    return respuesta

print(chat("¿Qué es la inteligencia artificial?"))
print(chat("Dame un ejemplo"))
print(chat("¿Qué es machine learning?"))
print(chat("¿Para qué sirve un dataset?"))
print(chat("¿Qué es una red neuronal?"))
```

Durante el desarrollo del chatbot, se observó que los modelos simples pueden generar respuestas imprecisas. Además, el manejo del contexto es limitado, lo que afecta la calidad de la conversación. También puede generar información incorrecta si no se controla. A pesar de esto, permite entender el funcionamiento básico de los chatbots.
